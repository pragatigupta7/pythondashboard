```python

total_sales = df['sales'].sum()
plt.hist(df['sales'], bins=10, edgecolor='black')
# Add title and labels
plt.title('Histogram')
plt.xlabel('sales')
plt.ylabel('price each')

# Show plot
plt.show()
```

```python
cities_list = df['city'].tolist()
sales_list = df['sales'].tolist()
# Group by city and sum the sales
city_sales = df.groupby('cities_list')['sales_list'].sum().reset_index()

# Display the city-wise sales
print(city_sales)

# Plot the bar graph for city-wise sales
plt.figure(figsize=(10, 6))
plt.bar(city_sales['cities_list'], city_sales['sales_list'], color='skyblue', edgecolor='black')

# Add title and labels
plt.title('City-wise Sales')
plt.xlabel('city')
plt.ylabel('Total Sales')

# Rotate city names for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
